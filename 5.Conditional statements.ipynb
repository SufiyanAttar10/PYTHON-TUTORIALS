## 1.2 if..elif..else statement

- In this case we have just one if and else statement but multiple elif statements .
- It checks conditions one by one until true condition found.
- If condition under if statement is false then it check conditions under elif statements and if all elif conditions are also false then it execute code under else statement.

```python
'''
here condition under if statement is true so code under if body get executed.
i.e. we found true condition under if statement so we can run code under that statement.
and no need to proceed further means no need to check further conditions.

'''
if (1<2): #True
    print("1 is less than 2 ")
    
elif (4>2): # true
    print("4 is greater than 2")
    
else:
    print("None of the above is correct ")
```

```python
'''
Here condition under if statement is false but under elif statement is true.
So as we found true condition under elif statement so we can run code under that statement.
and no need to proceed further means no need to check further conditions.
 
'''
if (1>2): # false
    print("1 is less than 2 ")
    
elif (4>2): # true
    print("4 is greater than 2")
    
else:
    print("None of the above is correct ")
```

```python
person = 'big data'

if person == 'python': # false
    print("eligible for a ds")
elif person == 'stats': # true
    print("eligible for a ds")
else:
    print("not eligible for a ds ")
```

```python
# here condition under if and elif statement are false so directly jump to else statement 
if (5<2): # false
    print("5 is less than 2 ")
elif (4<2): # false
    print("4 is less than 2")
else:
    print("None of the above is correct ")
```

## 1.3 Multiple conditional statements  

### Conditions hold by ' and '

In this case, if both conditions under if statements are true then and only then code under if statement get executed .

```python
# for ' and ' both condition must be true
if (5>3 or 5<10) and (3>4 or 6<8): 
    print("Correct")
else: 
    print("Incorrect")
```

```python
x = 6
y = 7
# for ' and ' both condition must be true
if x < y: 
    z = x + y
    print(z)
```

```python
# here first condition is true but second is not that's why code under if statement not getting executed
if 5>3 and 5<3 : 
    print("Correct")
else:
    print("Incorrect")
```

```python
# to check given number is multiple of 4 or 5 
num=int(input())

if num % 4 ==0 and num % 5 ==0:
    print("Given num is divisible by 4 and 5")
elif num % 4 == 0:
    print("Given num is divisible by 4")
elif num % 5 == 0:
    print("Given num is divisible by 5")
    
#num % 4 != 0 and num % 5 != 0
else:
    print("Given num is not divisible by 4 and 5 ")
```

## 1.4 nested if

```python
if 5>2: #parent condition
    print("we've entered into parent body")
    if 5>3: # child condition
        print("we've entered into child body") # body of child-if
```

```python
# check whether person is fully vaccinated or not using combination of if-elif-else and nested if

vaccine=str(input())
num_doses=int(input())

if vaccine =='covishield':
    if num_doses==2:
        print('Fully vaccinated')
    elif num_doses==1:
        print('Partially vaccinated')
    else:
        print('Not fully vaccinated')
```
