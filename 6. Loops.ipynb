## range() function 

- It is used to generate sequence of numbers .
- We can generate sequence by giving start and end point along with steps size. 
- Step size means difference between consecutive numbers of sequence.

#### Generate a sequence by giving arguments start and end point

```python
x = range(0,10)
list(x)
```

```python
# if we give start and end points then it generates the sequence from a start point to (ending point-1) with a step size 1 
# default step size is 1
print(list(range(0,10))) # range(0,m)..m --> m-1
```

#### Generate a sequence with step size

```python
# generate a sequence with step size
list(range(10, 20, 4))
```

```python
# Generate sequence with step size 6
print(list(range(0,20,6)))
```

#### Generate a sequence with end point

```python
# generate a sequence with end point
# default step size is 1
# if we gave only one point then it gives sequence from 0 to (that point -1) with step size 1 
print(list(range(5)))
```

## 1. for loop

- It is used to iterate over iterable objects such as String and Data structures.

```python
# create list of names
x = ['Ram', 'Sham', 'Kiran']
x
```

```python
for i in x:
    y = i*2
    print(y)
```

```python
for i in x:
    print(i)
    
print("Out of loop")
```

### for loop with else condition
- We can use else block in for loop . 
- It is executed after iterating over iterable object is done .
- If we add break condition in for loop then else statement it ingnored .

```python
# we can get sequence of squares 
for x in range(0,5):
    y = x * x
    print(y)
else:
    print("Iterating is over")
```

### nested for loop

```python
x = [1,2,3]
y = ['heena','ria','john']
for i in y:
    print('--------------body of parent')
    print(i)
    for j in x:
        print(j*j)
    print('--------------end of body of parent')
```

### iterate both list at a time - zip

```python
# if we use zip function then it will iterate upto the length of shorten list 
list1 = [1, 2, -3, 4, 7] # 5
list2 = [4, -6, 3, -1] # 4

for i,j in zip(list1,list2):
    print(i,j)
    print('add: ', i+j)
```

# 2. while loop
- It iterates continuously until condition become fails.
- It is very usefull when we don't know exactly how many times we have to iterates.



```python
i = 0 # initial value
while i<5: # while condition
    print(i) 
    i = i+1 # variable condition if it is not given then value of i remain zero and loops is remain as it is
```

### nested while loop

```python
# square of numbers upto 9
x = 0
y = 0
while x < 5 :
    while y < 5:
        print(x*y)
        x = x+1
        y = y+1
```

```python
# sum of same numbers upto 9
x = 0
y = 0
while x < 4 :
    while y < 4:
        z = x + y 
        print(z)
        x = x+1
        y = y+1
```

```python
# while and if
num = 15

# Condition of the while loop
while num > 10 :  # == 4
    # Find the mod of 2
    if num%2 == 0: 
        print("The number is even", num)
    else:
        print("The number is odd", num)
    # Increment num by 1
    num = num-1
```

# 3. break, continue, pass

- loops are continuosly iterating over code block until condition fails.
- but some times we want to stop iterating in this case we use break condition.
- but even if condition fails and we still want to iterate then we use continue.



```python
i = 0
while i<5:
    i = i+1 # here we first increment value of i 
    print(i) # and then updated value is printed that's why 0 is not printed
    if i==3: # breaking condition
        break
```

```python
# sum of same numbers upto 4
x = 0 # initilization
y = 0
while x < 5 : # parent 
    while y < 5: # child
        x = x+1 # increment 
        y = y+1 # increment
        print(x,y)
        # body of while loop
        z = x + y # 1+1 , 2+2
        if z < 5:
            print(z)
            break
```

```python
for num in range(2, 5):
    if num % 2 == 0:
        print("Found an even number", num)
        break
    print("Found an odd number", num)
```

```python
for num in range(2, 5):
    if num % 2 == 0:
        print("Found an even number", num)
        continue
    print("Found an odd number", num)
```
